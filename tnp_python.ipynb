```
def mergesort(b,p,r):
  if p < r:
    q = (p + r) // 2 # Use integer division
    mergesort(b, p, q)
    mergesort(b, q + 1, r)
    merge(b, p, q, r)

def merge(b,p,q,r):
  n1 = q - p + 1
  n2 = r - q

  L = [0] * n1  # Initialize with size n1
  R = [0] * n2  # Initialize with size n2

  for i in range(n1):
    L[i] = b[p + i]
  for j in range(n2):
    R[j] = b[q + 1 + j] # Corrected index for R

  i = 0
  j = 0
  k = p # Starting index for merged array

  while i < n1 and j < n2:
    if L[i] <= R[j]:
      b[k] = L[i]
      i += 1
    else:
      b[k] = R[j]
      j += 1
    k += 1

  while i < n1:
    b[k] = L[i]
    i += 1
    k += 1

  while j < n2:
    b[k] = R[j]
    j += 1
    k += 1

n=int(input("enter the size of the array: "))
b=[]
for i in range(0,n):
  print("enter the element at position ",i+1," :")
  e=int(input())
  b.append(e)
print(b)
p=0
r=n-1
print(p,r)
mergesort(b,p,r)
print(b)
```
